```python
import pandas as pd
import csv

```

```python
df = pd.read_csv('Linear Algebra Data - Sheet1.csv')
df.head()
```

```python
#Replace non values with zero
df['red_sig_str_percentage'].replace({'---':'0%'}, inplace=True)
df['blue_sig_str_percentage'].replace({'---':'0%'}, inplace=True)
df['red_takedown_percentage'].replace({'---':'0%'}, inplace=True)
df['red_takedown_percentage'].replace({'---':'0%'}, inplace=True)
```

```python
#add new columns
df.insert(14, 'red_total_strikes_percentage', '0%')
df.insert(15, 'blue_total_strikes_percentage', '0%')
```

```python
#iterate through
for index, row in df.iterrows():
    #get red_total_strikes value for row
    val = row['red_total_strikes']
    list = val.split(' ')

    #if no strikes were attempted
    if int(list[2]) != 0:
        df.at[index, 'red_total_strikes_percentage'] = f'{int(list[0])/int(list[2]):.0%}'
    
    #clean red_total_strikes
    df.at[index, 'red_total_strikes'] = list[0]

    #get red_total_strikes value for row
    val = row['blue_total_strikes']
    list = val.split(' ')

    #if no strikes were attempted
    if int(list[2]) != 0:
        df.at[index, 'blue_total_strikes_percentage'] = f'{int(list[0])/int(list[2]):.0%}'
    
    #clean blue_total_strikes
    df.at[index, 'blue_total_strikes'] = list[0]


    #removing inclusion of 'fractional' string for rest of columns that are displayed like this
    #clean significant strikes
    list = row['red_sig_str'].split(' ')
    df.at[index, 'red_sig_str'] = list[0]
    list = row['blue_sig_str'].split(' ')
    df.at[index, 'blue_sig_str'] = list[0]

    #clean takedowns
    list = row['red_takedowns'].split(' ')
    df.at[index, 'red_takedowns'] = list[0]
    list = row['blue_takedowns'].split(' ')
    df.at[index, 'blue_takedowns'] = list[0]

    


df.head()
```

```python
for index, row in df.iterrows():
    #change winner to 0 (red won) or 1 (blue won) or 2 for no winner
    if(row['winner'] == row['redCorner']):
        df.at[index, 'winner'] = 0
    if(row['winner'] == row['blueCorner']):
        df.at[index, 'winner'] = 1
    if pd.isna(row['winner']):
        df.at[index, 'winner'] = 2
df.head()
```
